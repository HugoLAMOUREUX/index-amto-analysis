```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ftplib import FTP
import csv
import io
from datetime import datetime, timedelta, date
```

# Get the data from the bucket

```python
ftp_url = os.getenv('FS_BUCKET_URL')
ftp_user = os.getenv('FS_BUCKET_USERNAME')
ftp_password = os.getenv('FS_BUCKET_PASSWORD')
```

```python
def get_ftp_content(date):
    if not ftp_url or not ftp_user or not ftp_password:
        print('The following env vars are required: FS_BUCKET_URL, FS_BUCKET_USERNAME and FS_BUCKET_PASSWORD')
        return None
    try:
        ftp = FTP(ftp_url)
        ftp.login(ftp_user, ftp_password)
        bio = io.BytesIO()
        filename = date
        ftp.retrbinary(f'RETR {filename}.csv', callback=bio.write)
        bio_value = bio.getvalue()
        decoded_content = bio_value.decode('UTF-8')
        return decoded_content
    except Exception as exception:
        print(exception)
        return None

def format_insee(raw_insee: str):
    if len(raw_insee) == 5 and raw_insee.isdigit():
        return raw_insee
    return None

def format_uv(raw_uv: str):  # '-999' or '' means no value
    parsed_uv = None
    if raw_uv.isdigit():
        parsed_uv = int(raw_uv)
        if parsed_uv < 0:
            parsed_uv = None
    return parsed_uv

def get_indices_from_date(date):
    decoded_content = get_ftp_content(date)
    if decoded_content is None:
        print("error decoded content")
    reader = csv.DictReader(
        decoded_content.splitlines(),
        delimiter=';'
    )
    indices = []
    
    for row in reader:
        insee = format_insee(row['insee'])
        indices.append({
            "zone_id": insee,
            "date": row['date'],
            "uv_j0": format_uv(row['UV_J0']),
            "uv_j1": format_uv(row['UV_J1']),
        })
    return indices
```

```python
datej0="20230705"
datej1="20230704"
j0=get_indices_from_date(datej0)
print(j0[0:4])
j1=get_indices_from_date(datej1)
print(j1[0:4])
```

# Analyze the data

```python
def calcul_pourcentage(date_j1):
    j0=get_indices_from_date((date_j1+timedelta(days=1)).strftime("%Y%m%d"))
    j1=get_indices_from_date(date_j1.strftime("%Y%m%d"))
    df0 = pd.DataFrame(j0)
    df1 = pd.DataFrame(j1)
    # df0.head()
    # df1.head()
    # print(df0["zone_id"].count(),"communes dans le df de J-0")
    # print(df1["zone_id"].count(),"communes dans le df de J-1")
    
    df = pd.merge(df0, df1, on=['zone_id'])
    df=df.drop(["uv_j1_x","uv_j0_y"],axis=1)
    df=df.rename(columns={"date_x":"date_j-0","date_y":"date_j-1","uv_j0_x":"j-0","uv_j1_y":"j-1"})
    # df = pd.concat([df0, df1], axis=1, join="inner")
    df.head()
    
    nb_total=df["zone_id"].count()
    df_dif=df[df["j-0"]!=df["j-1"]]
    nb_dif=df_dif["zone_id"].count()
    df_gap_1=df[abs(df["j-0"]-df["j-1"])>1]
    nb_gap_1=df_gap_1["zone_id"].count()
    
    pourcentage_gap_0=round(nb_dif/nb_total*100,2)
    pourcentage_gap_1=round(nb_gap_1/nb_total*100,2)
    # print(pourcentage_gap_0,"% des indices UV ont changés entre le",date_j1,"et le",date_j1+timedelta(days=1))
    # print(pourcentage_gap_1,"% des indices UV qui ont plus de 1 de différence entre le",date_j1,"et le",date_j1+timedelta(days=1))

    
    return {"gap_0":pourcentage_gap_0, "gap_1":pourcentage_gap_1}

date_test=date(2023,9,4)
calcul_pourcentage(date_test)
```

```python
date_debut = date(2022, 10, 1)

# Définissez la date de fin pour l'année 2022
date_fin = date(2022, 10, 31)

current_date = date_debut

indices_gap_0={}
indices_gap_1={}
while current_date <= date_fin:
    try:
        p=calcul_pourcentage(current_date)
        indices_gap_0[current_date.strftime("%Y%m%d")]=p["gap_0"]
        indices_gap_1[current_date.strftime("%Y%m%d")]=p["gap_1"]
        current_date += timedelta(days=1)
    except Exception as exception:
        print(exception)


print(indices_gap_0)
print(indices_gap_1)

```

```python
x_0=[indice for indice in indices_gap_0.keys() ]
y_0=[indice for indice in indices_gap_0.values()]
x_1=[indice for indice in indices_gap_1.keys() ]
y_1=[indice for indice in indices_gap_1.values()]

plt.plot(x_0, y_0,"r",label="% de différences entre J-0 et J-1")
plt.plot(x_1, y_1,"b",label="% de différence >1 entre J-0 et J-1")
plt.ylabel('% de différence entre J-0 et J-1 dans l\'indice UV')
plt.xticks(range(len(x_0)), x_0, rotation=60)
plt.legend(loc="upper right")
plt.show()

```
